```python
import pyproj

wgs84 = pyproj.Proj(projparams = 'epsg:4326')
InputGrid = pyproj.Proj(projparams = 'epsg:3035')

def unproject(coord1,coord2): #Allows to convert ESPG:3035 coordinates (projected) into ESPG:4326 coordinates (GPS)


    return pyproj.transform(InputGrid, wgs84, coord1, coord2)
```

```python
def swapPositions(list, pos1, pos2):
     
    list[pos1], list[pos2] = list[pos2], list[pos1]
    return list
```

```python
#The X_LLC and Y_LLC columns correspond to the projected coordinates of the lower left corner of each cells (ESPG:3035). We convert the GPS coordinates of the polygon into projected coordinates

coords_proj=[]

for k in range(len(coords)):
    coords_proj.append(pyproj.transform(wgs84, InputGrid, coords[k][0], coords[k][1]))

coords_proj
```

```python
#Swap positions of each tuple in the polygon list, for some reason they are not placed in the right order.

good_proj=[]
for k in range(len(coords)):
    tmp=swapPositions(list(coords_proj[k]),0,1)
    good_proj.append(tuple(tmp))

good_proj
```
